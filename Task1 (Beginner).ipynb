```python
import pandas as pd

data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'Salary': [50000, 60000, 70000]}

df = pd.DataFrame(data)
print(df)
```

```python
print("Mean Age:", df['Age'].mean())
print("Median Salary:", df['Salary'].median())
print("Standard Deviation of Age:", df['Age'].std())
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y, marker='o', linestyle='-')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Chart')
plt.show()
```

```python
plt.scatter(x, y, color='red')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')
plt.show()
```

```python
import seaborn as sns

sns.boxplot(df['Salary'])
plt.title('Salary Distribution')
plt.show()
```

```python
sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution with Density')
plt.show()
```

```python
dates = pd.date_range(start="2024-01-01", periods=10, freq='D')
time_series_data = pd.DataFrame({'Date': dates, 'Value': np.random.randint(10, 100, size=10)})
time_series_data.set_index('Date', inplace=True)

time_series_data.plot(title='Time Series Data')
plt.ylabel('Values')
plt.show()
```

```python
corr = df.select_dtypes(include=['number']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
grouped = df.groupby('Age')[df.select_dtypes(include='number').columns].mean()
print(grouped)
```

```python
df.loc[3] = [None, None, None]  # Adding a missing value row
print("Before filling missing values:")
print(df)

df.fillna(df.select_dtypes(include='number').mean(), inplace=True)
print("\nAfter filling missing values:")
print(df)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].plot(x, y, marker='o')
axs[0].set_title('Line Chart')

axs[1].bar(x, y, color='green')
axs[1].set_title('Bar Chart')

axs[2].scatter(x, y, color='red')
axs[2].set_title('Scatter Plot')

plt.tight_layout()
plt.show()
```

```python
plt.plot(x, y, marker='o', linestyle='--', color='purple', label='Custom Line')
plt.xlabel('Custom X-axis')
plt.ylabel('Custom Y-axis')
plt.title('Customized Line Chart')
plt.legend()
plt.grid(True)
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
print(titanic.head())
print("\nSummary Statistics:")
print(titanic.describe())
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='class', hue='survived', data=titanic)
plt.title('Survival Rate by Passenger Class')
plt.show()

sns.histplot(titanic['age'].dropna(), kde=True, bins=30)
plt.title('Age Distribution of Titanic Passengers')
plt.show()
```
